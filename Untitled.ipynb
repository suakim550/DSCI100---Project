# Data Description
The two datasets are from a data science project with real stakeholders, collected to predict the usage of a Minecraft research server.
## Player Data (players.csv)
* 7 columns and 197 rows
* Average value of hours played: 5.85
* Average value of age: 21.14
1. Experience:  “chr” type variable with five categories that represent the player’s experience level of Minecraft as “Beginner”, “Amateur”, “Regular”, “Veteran”, or “Pro”. 
2.	Subscribe: “lgl” type variable with binary responses, TRUE” or “FALSE”, that represents whether a player has subscribed to the game-related newsletter.
3.	hashedEmail: “chr” type variable that represents player’s encrypted email
4.	played_hours: “dbl” type variable that represents hours played
5.	Name: “chr” type variable that represents player’s name
6.	Gender: “chr” type variable that represents player’s gender with seven categories including “Female”, “Male”, “Agender”, “Non-binary”, “Two-Spirited”, “Other” and “Prefer not to say”. 
7.	Age: “dbl” type variable that represents player’s age.

  * Misreporting Data

# Questions
**Can hours played and age predict whether players have subscribed to the newsletter in players.csv?** The dataset players.csv will help address the question of interest, as it provides numerical information about each player’s age and hours played in the “dbl” type, and whether the player has subscribed to the game-related newsletter, as a binary response. Moreover, the sessions.csv dataset will not be used, as it doesn’t provide any helpful information for predicting subscription status.

4. Insights

```r
#compute the mean value:
average_played_hours <- players |>
    summarize(avg_played_hours = mean(played_hours, na.rm = TRUE))

average_age <- players |>
    summarize(avg_age = mean(Age, na.rm = TRUE))

average_played_hours
average_age

#Making plots

age_vs_played_hours <- ggplot(players, aes(x = Age, y = played_hours, color = subscribe)) +
                        geom_point(alpha = 0.7) +
                        labs(
                        title = "Age vs. Played Hours by Subscription Status",
                        x = "Age (years)",
                        y = "Hours Played") +
                        theme(text = element_text(size = 12)) +
                        scale_y_log10(labels = label_comma())
age_vs_played_hours 

```
